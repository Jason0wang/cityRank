```python
import pandas as pd
```

```python
newrank = pd.read_excel('./城市等级.xlsx',sheetname=1)
```

```python
oldrank = pd.read_excel('./城市等级.xlsx',sheetname=2)
```

```python
def trans(x):
    if "：" in x:
        #row['城市名称'] = temp.split("：")[1]
        return x.split("：")[0]
    else:
        return "直辖市"
```

```python
newrank['省份'] = newrank['城市名称'].map(trans)
```

```python
newrank['城市名称'] =  newrank['城市名称'].map(lambda x: x.split('：')[1] if '：'in x else x)
```

```python
oldrank = pd.merge(oldrank,newrank[['城市名称','省份']],on='城市名称',how='left')
```

```python
oldrank.to_excel('oldrank.xlsx',index=False)
```

```python
oldrank = oldrank.rename(columns = {'城市排名':'去年排名'})
```

```python
newrank = pd.merge(newrank,oldrank[['城市名称','去年排名']],on='城市名称',how='left')
```

```python
newrank['排名变动'] = -(newrank['排名'] - newrank['去年排名'])
```

```python
def updown(x):
    if x == 0:
        return "〓"
    elif x < 0:
        return "↓"+ str(-x)
    else:
        return "↑"+str(x)
# 可视化排名变动
newrank['排名变动'] = newrank['排名变动'].map(updown)
```

```python
#统计各省排名平均变动
total=dict()
count=dict()
for index,row in newrank.iterrows():
    if row['省份'] in total:
        total[row['省份']] += row['排名变动']
        count[row['省份']] += 1
    else:
        total[row['省份']] = row['排名变动']
        count[row['省份']] = 1
```

```python
avg=dict()
for i in total:
    avg[total[i]/count[i]] = i
```

```python
for i in sorted(avg.keys()):
    print (avg[i]+ " "+ str(i))
```
